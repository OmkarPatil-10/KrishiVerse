```python
import numpy as np
import pandas as pd
from random import randint
from datetime import timedelta
import warnings
warnings.filterwarnings('ignore')

# Load the original data
df_real = pd.read_csv('combined_dataset.csv')
print(f"Original total rows: {len(df_real)}")
print(f"Original districts: {df_real['District'].unique().tolist()}")

# All Maharashtra districts
all_districts = [
    'Ahmednagar', 'Akola', 'Amravati', 'Aurangabad', 'Beed', 'Bhandara',
    'Buldhana', 'Chandrapur', 'Dhule', 'Gadchiroli', 'Gondia', 'Hingoli',
    'Jalgaon', 'Kolhapur', 'Latur', 'Mumbai', 'Nagpur', 'Nanded', 'Nandurbar',
    'Nashik', 'Osmanabad', 'Palghar', 'Parbhani', 'Pune', 'Raigad', 'Ratnagiri',
    'Sangli', 'Satara', 'Sindhudurg', 'Solapur', 'Thane', 'Wardha', 'Washim', 'Yavatmal'
]

existing_districts = df_real['District'].unique().tolist()
# Fix: 'Amarawati' should be 'Amravati' - standardize the name
if 'Amarawati' in existing_districts:
    df_real['District'] = df_real['District'].replace('Amarawati', 'Amravati')
    existing_districts = df_real['District'].unique().tolist()

missing_districts = [d for d in all_districts if d not in existing_districts]

print(f"Existing districts: {len(existing_districts)} - {existing_districts}")
print(f"Missing districts: {len(missing_districts)}")

# Target total rows: 400k
TARGET_TOTAL_ROWS = 400000
rows_per_district = TARGET_TOTAL_ROWS // len(all_districts)
print(f"\nTarget rows per district: {rows_per_district}")
print(f"Target total rows: {rows_per_district * len(all_districts)}")

# First, analyze crop distribution in existing data
print("\n📊 Analyzing crop distribution in existing data...")

# Get unique commodities and their frequency in each existing district
commodity_distribution = {}
all_commodities = set()

for district in existing_districts:
    district_data = df_real[df_real['District'] == district]
    commodity_dist = district_data['Commodity'].value_counts(normalize=True)
    all_commodities.update(commodity_dist.index)
    
    # Get price statistics per commodity in each district
    price_stats = {}
    for commodity in district_data['Commodity'].unique():
        comm_data = district_data[district_data['Commodity'] == commodity]
        if len(comm_data) > 5:  # Only keep commodities with sufficient data
            price_stats[commodity] = {
                'mean_modal': comm_data['Modal_Price'].mean(),
                'std_modal': comm_data['Modal_Price'].std(),
                'mean_min': comm_data['Min_Price'].mean(),
                'mean_max': comm_data['Max_Price'].mean(),
                'q1_modal': comm_data['Modal_Price'].quantile(0.25),
                'q3_modal': comm_data['Modal_Price'].quantile(0.75),
                'min_modal': comm_data['Modal_Price'].min(),
                'max_modal': comm_data['Modal_Price'].max()
            }
    
    commodity_distribution[district] = {
        'distribution': commodity_dist.to_dict(),
        'price_stats': price_stats,
        'total_rows': len(district_data),
        'unique_crops': len(commodity_dist)
    }

print(f"✅ Analysis complete! Found {len(all_commodities)} unique commodities across all districts")

def get_similar_districts(target_district, existing_districts):
    """
    Enhanced function to find districts with similar agricultural patterns
    """
    # Enhanced crop zones with more detailed mapping
    crop_zones = {
        'Jalgaon': {
            'zone': 'Khandesh',
            'similar': ['Dhule', 'Nandurbar', 'Buldhana'],
            'major_crops': ['Banana', 'Cotton', 'Onion']
        },
        'Nashik': {
            'zone': 'North Maharashtra',
            'similar': ['Ahmednagar', 'Pune', 'Satara', 'Dhule'],
            'major_crops': ['Grapes', 'Onion', 'Tomato', 'Pomegranate']
        },
        'Pune': {
            'zone': 'Western Maharashtra',
            'similar': ['Satara', 'Sangli', 'Solapur', 'Ahmednagar'],
            'major_crops': ['Sugarcane', 'Grapes', 'Onion', 'Jowar']
        },
        'Kolhapur': {
            'zone': 'Southern Maharashtra',
            'similar': ['Sangli', 'Satara', 'Ratnagiri'],
            'major_crops': ['Sugarcane', 'Rice', 'Tur']
        },
        'Ratnagiri': {
            'zone': 'Konkan',
            'similar': ['Sindhudurg', 'Raigad', 'Thane'],
            'major_crops': ['Mango', 'Coconut', 'Rice', 'Cashew']
        },
        'Nagpur': {
            'zone': 'Vidarbha',
            'similar': ['Wardha', 'Amravati', 'Chandrapur', 'Bhandara'],
            'major_crops': ['Orange', 'Cotton', 'Rice', 'Tur']
        },
        'Amravati': {
            'zone': 'Vidarbha',
            'similar': ['Nagpur', 'Wardha', 'Buldhana', 'Akola'],
            'major_crops': ['Cotton', 'Soybean', 'Tur']
        },
        'Aurangabad': {
            'zone': 'Marathwada',
            'similar': ['Jalna', 'Beed', 'Parbhani', 'Nanded'],
            'major_crops': ['Cotton', 'Sugarcane', 'Onion', 'Tur']
        },
        'Mumbai': {
            'zone': 'Konkan',
            'similar': ['Thane', 'Raigad', 'Palghar'],
            'major_crops': ['Vegetables', 'Fruits', 'Rice']
        }
    }
    
    # Find matching zone
    similar = []
    for center, info in crop_zones.items():
        if target_district == center or target_district in info['similar']:
            # Add center and all similar districts that exist in our data
            candidates = [center] + info['similar']
            similar = [d for d in candidates if d in existing_districts]
            break
    
    # If no zone match, find districts with similar characteristics
    if not similar:
        # Use geographic proximity (simplified - you can enhance with actual coordinates)
        coastal = ['Mumbai', 'Raigad', 'Ratnagiri', 'Sindhudurg', 'Thane', 'Palghar']
        western = ['Pune', 'Satara', 'Sangli', 'Solapur', 'Kolhapur']
        vidarbha = ['Nagpur', 'Amravati', 'Wardha', 'Chandrapur', 'Gondia']
        marathwada = ['Aurangabad', 'Beed', 'Nanded', 'Parbhani', 'Latur', 'Osmanabad', 'Hingoli']
        khandesh = ['Jalgaon', 'Dhule', 'Nandurbar']
        
        if target_district in coastal:
            similar = [d for d in coastal if d in existing_districts]
        elif target_district in western:
            similar = [d for d in western if d in existing_districts]
        elif target_district in vidarbha:
            similar = [d for d in vidarbha if d in existing_districts]
        elif target_district in marathwada:
            similar = [d for d in marathwada if d in existing_districts]
        elif target_district in khandesh:
            similar = [d for d in khandesh if d in existing_districts]
    
    # Return top 3 similar districts or default to all existing
    return similar[:3] if similar else existing_districts[:3]

def create_agriculturally_aware_synthetic_data(df_real, missing_districts, existing_districts, rows_per_district):
    """
    Create synthetic data that preserves realistic crop distributions
    """
    
    # Pre-process existing data
    print("\n🔄 Pre-processing existing district data...")
    district_data = {}
    for district in existing_districts:
        district_data[district] = df_real[df_real['District'] == district].copy()
        print(f"  {district}: {len(district_data[district])} rows, {district_data[district]['Commodity'].nunique()} crops")
    
    # Pre-process all unique varieties and grades to maintain consistency
    variety_map = {}
    grade_map = {}
    for district in existing_districts:
        for _, row in district_data[district].iterrows():
            key = (row['Commodity'], district)
            if key not in variety_map:
                variety_map[key] = row['Variety']
                grade_map[key] = row['Grade']
    
    synthetic_rows = []
    
    print("\n🌾 Generating agriculturally-aware synthetic data...")
    
    for idx, new_district in enumerate(missing_districts):
        print(f"\nProcessing {idx+1}/{len(missing_districts)}: {new_district}")
        
        # Find similar districts based on geography/agriculture
        similar_districts = get_similar_districts(new_district, existing_districts)
        print(f"  Similar districts: {similar_districts}")
        
        # Combine data from similar districts to learn crop distribution
        combined_crop_dist = {}
        combined_price_stats = {}
        
        for district in similar_districts:
            if district in commodity_distribution:
                # Weight by how similar the district is (first district gets highest weight)
                weight = 1.0 / (similar_districts.index(district) + 1)
                
                for crop, prob in commodity_distribution[district]['distribution'].items():
                    combined_crop_dist[crop] = combined_crop_dist.get(crop, 0) + prob * weight
                
                for crop, stats in commodity_distribution[district]['price_stats'].items():
                    if crop not in combined_price_stats:
                        combined_price_stats[crop] = {
                            'count': 0, 
                            'mean_modal': 0, 
                            'mean_min': 0, 
                            'mean_max': 0,
                            'q1_modal': 0,
                            'q3_modal': 0
                        }
                    combined_price_stats[crop]['count'] += weight
                    combined_price_stats[crop]['mean_modal'] += stats['mean_modal'] * weight
                    combined_price_stats[crop]['mean_min'] += stats['mean_min'] * weight
                    combined_price_stats[crop]['mean_max'] += stats['mean_max'] * weight
        
        # Normalize
        total_weight = sum(combined_crop_dist.values())
        for crop in combined_crop_dist:
            combined_crop_dist[crop] /= total_weight
        
        for crop in combined_price_stats:
            count = combined_price_stats[crop]['count']
            if count > 0:
                combined_price_stats[crop]['mean_modal'] /= count
                combined_price_stats[crop]['mean_min'] /= count
                combined_price_stats[crop]['mean_max'] /= count
        
        # Generate rows based on this distribution
        crops = list(combined_crop_dist.keys())
        crop_probs = list(combined_crop_dist.values())
        
        # Select crops for all rows
        selected_crops = np.random.choice(crops, size=rows_per_district, p=crop_probs)
        
        district_synthetic = []
        
        # Process in batches for better performance
        batch_size = 1000
        for i in range(0, len(selected_crops), batch_size):
            batch_crops = selected_crops[i:i+batch_size]
            
            for crop in batch_crops:
                # Find a parent district that has this crop
                parent_found = False
                for parent in similar_districts:
                    if parent in district_data and crop in district_data[parent]['Commodity'].values:
                        crop_rows = district_data[parent][district_data[parent]['Commodity'] == crop]
                        if len(crop_rows) > 0:
                            parent_row = crop_rows.sample(1).iloc[0].to_dict()
                            parent_found = True
                            break
                
                if not parent_found:
                    # Try all existing districts
                    for parent in existing_districts:
                        if parent in district_data and crop in district_data[parent]['Commodity'].values:
                            crop_rows = district_data[parent][district_data[parent]['Commodity'] == crop]
                            if len(crop_rows) > 0:
                                parent_row = crop_rows.sample(1).iloc[0].to_dict()
                                parent_found = True
                                break
                
                if not parent_found:
                    # Last resort: get any row from the most similar district
                    parent_row = district_data[similar_districts[0]].sample(1).iloc[0].to_dict()
                
                # Apply realistic price variations
                if crop in combined_price_stats:
                    stats = combined_price_stats[crop]
                    # Use normal distribution based on actual statistics
                    variation = np.random.normal(1, 0.08)  # 8% variation
                    
                    parent_row['Modal_Price'] = max(1, stats['mean_modal'] * variation)
                    parent_row['Min_Price'] = max(1, stats['mean_min'] * variation)
                    parent_row['Max_Price'] = max(1, stats['mean_max'] * variation)
                else:
                    price_factor = np.random.normal(1, 0.1)
                    parent_row['Modal_Price'] = max(1, parent_row['Modal_Price'] * price_factor)
                    parent_row['Min_Price'] = max(1, parent_row['Min_Price'] * price_factor)
                    parent_row['Max_Price'] = max(1, parent_row['Max_Price'] * price_factor)
                
                # Ensure min < modal < max
                prices = sorted([parent_row['Min_Price'], parent_row['Modal_Price'], parent_row['Max_Price']])
                parent_row['Min_Price'], parent_row['Modal_Price'], parent_row['Max_Price'] = prices
                
                # Date variation
                if 'Arrival_Date' in parent_row:
                    try:
                        if isinstance(parent_row['Arrival_Date'], str):
                            date_val = pd.to_datetime(parent_row['Arrival_Date'], dayfirst=True)
                        else:
                            date_val = parent_row['Arrival_Date']
                        date_offset = randint(-5, 5)
                        parent_row['Arrival_Date'] = date_val + timedelta(days=date_offset)
                    except:
                        parent_row['Arrival_Date'] = pd.Timestamp.now()
                
                # Market modification
                if 'Market' in parent_row:
                    parent_row['Market'] = f"{parent_row['Market']}_{new_district[:3]}"
                
                # Set new district
                parent_row['District'] = new_district
                
                # Round prices
                for col in ['Min_Price', 'Max_Price', 'Modal_Price']:
                    if col in parent_row:
                        parent_row[col] = round(parent_row[col], 2)
                
                district_synthetic.append(parent_row)
        
        synthetic_rows.extend(district_synthetic)
        
        # Print statistics for this district
        df_temp = pd.DataFrame(district_synthetic)
        unique_crops = df_temp['Commodity'].value_counts()
        print(f"  ✓ Generated {len(district_synthetic)} rows")
        print(f"    Crops: {len(unique_crops)} unique")
        print(f"    Top 3: {', '.join(unique_crops.index[:3])}")
        print(f"    Avg price: ₹{df_temp['Modal_Price'].mean():.2f}")
    
    return pd.DataFrame(synthetic_rows)

# Generate synthetic data with agricultural awareness
df_synthetic = create_agriculturally_aware_synthetic_data(
    df_real, missing_districts, existing_districts, rows_per_district
)

# Combine original and synthetic data
df_all = pd.concat([df_real, df_synthetic], ignore_index=True)

print(f"\n{'='*50}")
print(f"✅ FINAL DATASET CREATED!")
print(f"{'='*50}")
print(f"Total rows: {len(df_all):,}")
print(f"Districts covered: {df_all['District'].nunique()}/{len(all_districts)}")
print(f"Unique commodities: {df_all['Commodity'].nunique():,}")

# Verify crop distribution
print("\n🌾 Top 20 commodities in final dataset:")
print(df_all['Commodity'].value_counts().head(20))

print("\n📊 Distribution by district:")
district_counts = df_all['District'].value_counts().sort_index()
for district, count in district_counts.items():
    print(f"  {district}: {count:,} rows")

# Save the dataset
df_all.to_csv('maharashtra_districts_complete.csv', index=False)
print("\n✅ Saved to 'maharashtra_districts_complete.csv'")

# Also save a sample for quick verification
df_all.sample(min(10000, len(df_all))).to_csv('maharashtra_districts_sample.csv', index=False)
print("✅ Saved sample to 'maharashtra_districts_sample.csv'")
```
